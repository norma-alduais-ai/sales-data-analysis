```python

import matplotlib.pyplot as plt  
import seaborn as sns  

# to work with data
import pandas as pd  

df = pd.read_csv("sales_data_dirty.csv")  
```

```python
#each deal for Sales_Amount scope
plt.figure(figsize=(10,6))
sns.histplot(df['Sales_Amount'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Sales Amount")
plt.xlabel("Sales Amount")
plt.ylabel("Number of Sales")
plt.show()
```

```python
#Boxplot to discover for outliers for sales 
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Sales_Amount'], color='lightgreen')
plt.title("Boxplot of Sales Amount")
plt.show()
```

```python
# Scatterplot: reation bet price and sales quantity 
plt.figure(figsize=(10,6))
sns.scatterplot(x='Unit_Price', y='Quantity_Sold', data=df, hue='Product_Category')
plt.title("Unit Price vs Quantity Sold by Product Category")
plt.xlabel("Unit Price")
plt.ylabel("Quantity Sold")
plt.legend(title='Product Category')
plt.show()
 
```

```python
#Barplot: sales by resion  
plt.figure(figsize=(8,6))
sns.barplot(x='Region', y='Sales_Amount', data=df, estimator=sum, ci=None, palette='pastel')
plt.title("Total Sales by Region")
plt.ylabel("Total Sales Amount")
plt.show()
 
```

```python
#Scatter + Regression Line: relation bet discount and sales 
plt.figure(figsize=(10,6))
sns.regplot(x='Discount', y='Sales_Amount', data=df, scatter_kws={'color':'orange'}, line_kws={'color':'red'})
plt.title("Sales Amount vs Discount")
plt.xlabel("Discount")
plt.ylabel("Sales Amount")
plt.show()
 
```

```python
#sales depens on product category 
plt.figure(figsize=(10,6))
sns.barplot(x='Product_Category', y='Sales_Amount', data=df, estimator=sum, ci=None)
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales Amount")
plt.show()
```

```python
#sales depends clients type 
plt.figure(figsize=(6,5))
sns.barplot(x='Customer_Type', y='Sales_Amount', data=df, estimator='mean', ci=None)
plt.title("Average Sales by Customer Type")
plt.ylabel("Average Sales Amount")
plt.show()
```

```python
#sales depands on sale chanel 
plt.figure(figsize=(6,5))
sns.boxplot(x='Sales_Channel', y='Sales_Amount', data=df)
plt.title("Sales Distribution by Channel")
plt.show()
```
